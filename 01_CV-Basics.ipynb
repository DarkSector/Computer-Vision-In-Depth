# CV Basics

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = "./images/lenna.png"
```

## Simple read and show 

```python
img = cv2.imread(image)
```

```python
# CV2 reads in BGR, we need to see it RGB
# https://stackoverflow.com/questions/46236180/opencv-imshow-will-cause-jupyter-notebook-crash
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Indexing an image

```python
# image is 512x512x3 and is a simple numpy array
img.shape
```

```python
# We want 200th row and 300th column to show a black rectangle of color black. Rectangle should have 100 x 100 pixels
img2 = img
img2[199:299, 299:399, :] = 0
```

```python
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# get center of image
w = img.shape[0]
h = img.shape[1]
(cX, cY) = (w // 2, h // 2)
```

```python
tl = img[0:cY, 0:cX]
plt.imshow(cv2.cvtColor(tl, cv2.COLOR_BGR2RGB))
plt.show()
```

## Shapes and stuff

```python
canvas = np.zeros((300, 300, 3), dtype="uint8")

green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
plt.imshow(canvas)
plt.show()
```

```python
canvas = np.zeros((300, 300, 3), dtype="uint8")

for i in range(0, 25):
	# randomly generate a radius size between 5 and 200, generate a random
	# color, and then pick a random point on our canvas where the circle
	# will be drawn
	radius = np.random.randint(5, high=200)
	color = np.random.randint(0, high=256, size = (3,)).tolist()
	pt = np.random.randint(0, high=300, size = (2,))
	
    # draw our random circle
	cv2.circle(canvas, tuple(pt), radius, color, -1)

# Show our masterpiece
plt.imshow(canvas)
plt.show()
```
